# Speech-to-text

## Separating different speakers

```python
from google.cloud import speech_v1p1beta1 as speech
import os
from pydub import AudioSegment
```

```python
sound = AudioSegment.from_mp3("input.mp3")
sound.export("output.wav", format="wav")
```

```python
sound = AudioSegment.from_wav("output.wav")
sound = sound.set_channels(1)
sound.export("output.wav", format="wav")
```

```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]=r'C:\Users\EPEAFF1\Desktop\Tech\springboard\key.json'
```

```python
client = speech.SpeechClient()
```

```python
speech_file = 'output.wav'
```

```python
with open(speech_file, 'rb') as audio_file:
    content = audio_file.read()
```

```python
audio = speech.types.RecognitionAudio(content=content)
```

```python
config = speech.types.RecognitionConfig(
    encoding=speech.enums.RecognitionConfig.AudioEncoding.LINEAR16,
    #sample_rate_hertz=8000,
    language_code='en-US',
    enable_speaker_diarization=True,
    diarization_speaker_count=2)
```

```python
print('Waiting for operation to complete...')
response = client.recognize(config, audio)
```

```python
# The transcript within each result is separate and sequential per result.
# However, the words list within an alternative includes all the words
# from all the results thus far. Thus, to get all the words with speaker
# tags, you only have to take the words list from the last result:
result = response.results[-1]
```

```python
words_info = result.alternatives[0].words
words_info
```

```python
# Printing out the output:
phrase = []
speaker = '1'
for word_info in words_info:
    if phrase==[] or (speaker == word_info.speaker_tag):
        phrase.append(word_info.word)
    else:
        print('Speaker {}: "{}"'.format(word_info.speaker_tag, ' '.join(phrase)))
        phrase = [word_info.word]
        speaker = word_info.speaker_tag
```
